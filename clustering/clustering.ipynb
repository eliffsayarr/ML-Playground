Exploring and Importing Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
data = pd.read_csv('tripadvisor_review.csv', sep=',')
data.info()
```

```python
column_names = ['user_id', 'art_galleries', 'dance_clubs', 'juice_bars',
                'restaurants', 'museums', 'resorts', 'park_picnic_spots',
                'beaches', 'theaters', 'religious_institutions'
                ]

data.columns = column_names
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.drop('user_id', axis = 1, inplace = True)
```

Using Robust Scalling to Handle Outliers

PCA to Lower the number of dimensions

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
pca = PCA(n_components = 2)
pca_1 = pca.fit_transform(data)
pca_1
```

```python
pca.components_
```

```python
new_data = pd.DataFrame({'PCA_1': pca_1[:, 0], 'PCA_2': pca_1[:, 1]})
new_data.head()
```

```python
cluster_test = plt.subplots(figsize=(12, 12))

cluster_test = sns.scatterplot(x = 'PCA_1', y = 'PCA_2', data = new_data, color = '#9b59b6')


sns.despine()
cluster_test.set_xlabel('PCA 1', fontsize = 15)
cluster_test.set_ylabel('PCA 1', fontsize = 15)
cluster_test.set_title('Preclustering', fontsize = 20)
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4)
```

```python
kmeans.fit_transform(new_data)
```

```python
new_data['cluster'] = kmeans.labels_
```

```python
cluster = plt.subplots(figsize=(20, 20))
cluster = sns.scatterplot(x = 'PCA_1', y = 'PCA_2', data = new_data, hue = 'cluster')
cluster.scatter(kmeans.cluster_centers_[:,0],
           kmeans.cluster_centers_[:,1], c = '#9b59b6', s = 1000, marker = 'o')

coeff = np.transpose(pca.components_[0:2, :])
n = coeff.shape[0]
columns = list(data.columns)

for i in range(n):
    plt.text(coeff[i,0]*3, coeff[i,1]*3, columns[i], size = 19,color = '#00586e')
    
    
sns.despine()
cluster.set_xlabel('PCA 1', fontsize = 15)
cluster.set_ylabel('PCA 1', fontsize = 15)
cluster.set_title('Clustering', fontsize = 20)
plt.show()
```
